```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# Generating the Lorentz-like spectrum

```python
x = np.arange(-3.0,3.0, 0.05)
signal_true = 1.0 / (1.0 + x**2.0 / 0.2)
signal_observed = 1.0 / (1.0 + x**2.0 / 0.2) + np.random.randn(x.shape[0])*0.05
```

```python
plt.plot(x, signal_observed)
```

## ファイルに保存

```python
import os
dirname = "tutorial3_files"
if not os.path.exists(dirname):
    os.mkdir(dirname)
```

```python
np.savetxt('tutorial3_files/x_lorentz.txt', x, delimiter=',')
np.savetxt('tutorial3_files/y_lorentz.txt', signal_observed, delimiter=',')
```
